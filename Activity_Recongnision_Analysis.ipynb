```python
##import the data from the files

import os
from os import path
import pandas as pd
import numpy as np
import pickle
from pandas_profiling import ProfileReport
from sklearn.model_selection import train_test_split
```

```python
path = "F:/"

S1_Dataset_Path = os.path.join(path, "Assignments", "S1_Dataset" )
S2_Dataset_Path = os.path.join(path, "Assignments", "S2_Dataset")

```

```python


path = "F:/"

## Define the values for gender
## Male = 0 and Female = 1 . Gender needs to be added in the last column.



S1_Dataset_Path = os.path.join(path, "Assignments", "S1_Dataset" )
S2_Dataset_Path = os.path.join(path, "Assignments", "S2_Dataset")

s1_data = pd.DataFrame()

for dataset in [S1_Dataset_Path]:
    for dataset in [S2_Dataset_Path]:
        data_path = os.path.join(path,dataset)

for filename in os.listdir(data_path):
       if filename != "README.txt" : 
            file_path = os.path.join(data_path,filename)
        
       ## if filename[-1] == 'F':
         ##   gender = 1
            
        ## else:
           ## gender = 0
            
data = pd.read_csv(file_path, header=None)
s1_data = s1_data.append(data, ignore_index=True)
        
        
  
    
s1_data.columns = ['Time', 'Frontal_Acc', 'Vertical_Acc', 'Lateral_Acc', 'ID', 'RSSI', 'Phase', ' Frequency', 'Activity' ]

print("Updated successfully!!! Done!!")

s1_data.info()
```
